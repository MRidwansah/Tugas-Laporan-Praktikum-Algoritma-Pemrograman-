Writing comments in python

Selain menulis kode, perhatikan bahwa menambahkan komentar ke kode Anda adalah ide yang bagus. Ini akan membantu orang lain memahami apa yang ingin Anda capai (alasan mengapa Anda menulis cuplikan kode tertentu). Hal ini tidak hanya membantu orang lain memahami kode Anda, tetapi juga dapat berfungsi sebagai pengingat bagi Anda ketika Anda kembali menggunakannya beberapa minggu atau bulan kemudian.

    Untuk menulis komentar dengan Python, gunakan simbol angka <code>#</code> sebelum menulis komentar Anda. Saat Anda menjalankan kode, Python akan mengabaikan semua yang melewati <code>#</code> pada baris tertentu.

```python
# Practice on writing comments

print('Hello, Python!') # This line prints a string
print('Hi') #untuk print Hi
```

<p>
    Setelah mengeksekusi sel di atas, Anda akan melihat bahwa <code>Baris ini mencetak string</code> tidak muncul di output, karena itu adalah komentar (dan diabaikan oleh Python).
</p>
<p>
    Baris kedua juga tidak dieksekusi karena <code>print('Hi')</code> juga diawali dengan tanda angka (<code>#</code>)! Karena ini bukan komentar penjelasan dari pemrogram, namun merupakan baris kode sebenarnya, kita dapat mengatakan bahwa pemrogram <em>mengomentari</em> baris kode kedua tersebut.

## <h3 id="errors">Errors in Python</h3>

<p>Setiap orang pasti pernah melakukan kesalahan. Untuk banyak jenis kesalahan, Python akan memberi tahu Anda bahwa Anda telah melakukan kesalahan dengan memberikan pesan kesalahan. Penting untuk membaca pesan kesalahan dengan cermat untuk benar-benar memahami di mana Anda melakukan kesalahan dan bagaimana Anda dapat memperbaikinya.</p>
<p>Misalnya, jika Anda mengeja <code>print</code> sebagai <code>frint</code>, Python akan menampilkan pesan kesalahan. Cobalah:</p>

```python
# Print string as error messsage

frint("Hello, Python!")
```

<p>Pesan kesalahan memberitahukan Anda:
<ol>
    <li>tempat terjadinya kesalahan (lebih berguna pada sel atau skrip buku catatan besar), dan</li>
    <li>jenis kesalahan apa itu (NameError)</li>
</ol>
<p>Di sini, Python mencoba menjalankan fungsi <code>frint</code>, tetapi tidak dapat menentukan apa itu <code>frint</code> karena ini bukan fungsi bawaan dan belum ditentukan sebelumnya oleh kami juga.</p>

hal>
    Anda akan melihat bahwa jika kita membuat jenis kesalahan yang berbeda, dengan lupa menutup string, kita akan mendapatkan kesalahan yang berbeda (yaitu, <code>SyntaxError</code>). Cobalah di bawah ini:
</p>

## <h3 id="python_error">Does Python kbow about your error before it runs your code?</h3>

Python adalah apa yang disebut <em>bahasa yang ditafsirkan</em>. Bahasa yang dikompilasi memeriksa seluruh program Anda pada waktu kompilasi, dan dapat memperingatkan Anda tentang seluruh kelas kesalahan sebelum eksekusi. Sebaliknya, Python menafsirkan skrip Anda baris demi baris saat dijalankan. Python akan berhenti mengeksekusi seluruh program ketika menemui kesalahan (kecuali kesalahan tersebut diharapkan dan ditangani oleh pemrogram, subjek lebih lanjut yang akan kita bahas nanti dalam kursus ini).

Coba jalankan kode pada sel di bawah ini dan lihat apa yang terjadi:

```python
# Print string and error to see the running order

print("This will be printed")
frint("This will cause an error")
print("This will NOT be printed")
```
